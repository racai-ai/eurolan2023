# Rulare transformari

https://github.com/GEM-benchmark/NL-Augmenter/tree/main/nlaugmenter/transformations


```python
from nlaugmenter.transformations.butter_fingers_perturbation.transformation import ButterFingersPerturbation
t1 = ButterFingersPerturbation(max_outputs=3)
t1.generate("Jason wants to move back to India by the end of next year.")
```

```python
t1.generate("Gheorghe a fost în concediu la Mamaia.")
```

```python
s1=["Gheorghe", "Ion", "a", "fost", "în", "concediu", "la", "Mamaia", "."]
l1=["B-PER",   "I-PER","O", "O",    "O",  "O",        "O",  "B-LOC",  "O"]
```

```python
from nlaugmenter.transformations.longer_names_ner.transformation import LongerNamesNer
t2 = LongerNamesNer(max_outputs=2)
r=t2.generate(s1,l1)
for (a,b) in r:
  print(a)
  print(b)
```

```python
from nlaugmenter.transformations.longer_location_ner.transformation import LongerLocationNer
t3 = LongerLocationNer(max_outputs=2)
r=t3.generate(s1,l1)
for (a,b) in r:
  print(a)
  print(b)
```

```python
from nlaugmenter.transformations.add_hashtags.transformation import HashtagGeneration
t4 = HashtagGeneration()
t4.generate(" ".join(s1))
```

```python
from nlaugmenter.transformations.geo_names_transformation.transformation import GeoNamesTransformation
t5 = GeoNamesTransformation()
t5.generate("Gheorghe visited Budapest during his vacation.")
```

```python
from nlaugmenter.transformations.neural_question_paraphraser.transformation import NeuralParaphaserPerturbation
t6 = NeuralParaphaserPerturbation(max_output=5)

```

```python
t6.generate("Where was Gheorghe during his vacation?")
```

```python
t6.generate("What city did he visit?")
```

```python
t6.generate("Did John visit Brașov on his trip to Romania, last year?")
```

```python
from nlaugmenter.transformations.contextual_meaning_perturbation.transformation import ContextualMeaningPerturbation
t7 = ContextualMeaningPerturbation(max_outputs=5)
```

```python
t7.generate("During his vacation, John visited Brașov, a very nice city in Romania.")
```

```python
t7 = ContextualMeaningPerturbation(max_outputs=5, perturbation_rate=0.5)
t7.generate("Brașov is a city situated in the central part of Romania.")
```
